```python
import nltk
def calculate_tag_score_QA(question, response):
    nounsQ = []
    verbsQ = []
    adjQ = []
    advQ = []
    for word, pos in nltk.pos_tag(nltk.word_tokenize(str(question))):
        if (pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS'):
            nounsQ.append(word)
        elif (pos == 'VB' or pos == 'VBD' or pos == 'VBG' or pos == 'VBN' or pos == 'VBP' or pos == 'VBZ'):
            verbsQ.append(word)
        elif (pos == 'JJ' or pos == 'JJR' or pos == 'JJS'):
            adjQ.append(word)
        elif (pos == 'RB' or pos == 'RBR' or pos == 'RBS'):
            advQ.append(word)
    
    nounsQ = set(nounsQ)
    nounsQ = list(set(nounsQ))
   
    verbsQ = set(verbsQ)
    verbsQ = list(set(verbsQ))
    
    adjQ = set(adjQ)
    adjQ = list(set(adjQ))
    
    advQ = set(advQ)
    advQ = list(set(advQ))
    
    len_max_count_tag = max(len(nounsQ), len(verbsQ), len(adjQ), len(advQ))
#     print(len_max_count_tag)
    tags = [nounsQ, verbsQ, adjQ, advQ]
    tag_count = 0
    max_tags = ""

    for tag in tags:
        tag_count +=1
        if len_max_count_tag == len(tag):
            max_count_tag = tag
            which_tag = tag_count
#     print(max_count_tag)

    if which_tag == 1:
        max_tags = "Noun"
    elif which_tag == 2:
        max_tags = "Verb"
    elif which_tag == 3:
        max_tags = "Adj"
    elif which_tag == 4:
        max_tags = "Adv"

#     print(max_tags)
    # nounsA = []
    # verbsA = []
    # adjA = []
    # advA = []

    tags_answer = []
    for word, pos in nltk.pos_tag(nltk.word_tokenize(str(response))):
        if max_tags == "Noun":
            if (pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS'):
                tags_answer.append(word)
        elif max_tags == "Verb":
            if (pos == 'VB' or pos == 'VBD' or pos == 'VBG' or pos == 'VBN' or pos == 'VBP' or pos == 'VBZ'):
                tags_answer.append(word)
        elif max_tags == "Adj":
            if (pos == 'JJ' or pos == 'JJR' or pos == 'JJS'):
                tags_answer.append(word)
        elif max_tags == "Adv":
            if (pos == 'RB' or pos == 'RBR' or pos == 'RBS'):
                tags_answer.append(word)

    tags_answer = set(tags_answer)
    tags_answer = list(set(tags_answer))
#     print(tags_answer)
    max_count_tag = set(max_count_tag)
    tags_answer = set(tags_answer)
    intersection_QA = max_count_tag.intersection(tags_answer)
    ratioQA = len(intersection_QA)/len_max_count_tag
    print(ratioQA)
    return ratioQA
```
